```python
import numpy
import matplotlib.pyplot as plt

from sscResolution import frc, fsc, statistics

FS  = [7,7]
FoS = [12,12]
```

# (1) FRCorrelation from two images

## (a) Computing curves

```python
from scipy.ndimage import gaussian_filter
from scipy import interpolate

def zoom(img,a):
    xx = numpy.linspace(-1.0,1.0,max(img.shape[1], img.shape[0]) )
    yy = numpy.linspace(-1.0,1.0,max(img.shape[1], img.shape[0]))
    fun =  interpolate.RectBivariateSpline(xx, yy, img, kx=1, ky=1, s=0)
    x = numpy.linspace(-a,a,max(img.shape[1], img.shape[0]))
    y = numpy.linspace(-a,a,max(img.shape[1], img.shape[0]))
    return fun(x,y)

r = 40
f = numpy.zeros([512,512])
f[256-r:256+r,256-r:256+r] = 10

sig = 20
f1  = gaussian_filter(f, sig)

fig = plt.figure(figsize=(20,20))

ax1 = fig.add_subplot(121)
ax1.imshow(f)
ax1.axis('off')
ax1.set_title(r'$f$')

ax2 = fig.add_subplot(122)
ax2.imshow(f1)
ax2.axis('off')
ax2.set_title(r'$f_1$')

dic = frc.compute2i( f, f1 )    
```

## (b) FRC plot 

```python
#full plot
frc.plot(dic, {'label': r'$c(f,f_1)$','unit':"um", "pxlsize": 1, 'figsize': FS, 'fontsize':FoS, 'full': True} )

#simplified plot
frc.plot(dic, {'label': r'$c(f,f_1)$','unit':"um", "pxlsize": 1, 'figsize': FS, 'fontsize':FoS } )
```

## (c) Get curves

```python
C = dic['curves']['C'].real  #correlation curve
H = dic['curves']['H']       #halfbit curve
S = dic['curves']['S']       #3-sigma curve
B = dic['curves']['H']       #Fisher-Niculescu curve
```

## (d) Get resolution numbers

```python
xH = dic['x']['H']  #intersect {C, H}
xS = dic['x']['S']  #intersect {C, S}
xB = dic['x']['S']  #intersect {C, B}

print('x-halfbit:', xH)
print('x-sigma  :', xS)
print('x-fisher :', xB)
```

# (3) 2D FRC Map

```python
image = numpy.load('glassphere.npy')

params = {'threshold':'fisher','ngrid':120,'L':[40,0.8],'eps':40,'running':frc.SERIAL,'shape':image.shape }

usr = {'label':'Fisher','figsize': FS,'fontsize':FoS}

rmap = frc.map(image, params)

plt.figure(figsize=FS)
plt.imshow(rmap)

statistics.histmap( rmap, params, usr )
```

# (4) FSC for a single volume

## (a) Generic volume

```python
n = 512
t = numpy.linspace(-1,1,n)

L = [40, 0.3]

def sigmoidal(t,L):
    return 1.0/(1 + numpy.exp(-L[0]*(t+L[1]))) - 1.0/(1 + numpy.exp(-L[0]*(t-L[1])))

u,v,w = numpy.meshgrid(t,t,t)
vol   = sigmoidal(u,L) * sigmoidal(v,L) * sigmoidal(w,L)
```

```python
plt.imshow( numpy.hstack((vol[256,:,:],vol[:,256,:],vol[:,:,256])))
```

## (b) Computing FSC

```python
dic = fsc.compute(vol)

fsc.plot(dic, {'label': "Sample", 'unit': "nm", "pxlsize": 1, 'figsize':FS, 'full': True} )
```

# (5) 3D FRC Map

```python
params = {'ngrid':30,'L':[20,0.5],'eps':40,'running':frc.SERIAL,'threshold':'fisher', 'shape':vol.shape}

vmap = fsc.map(vol, params)
```

```python
plt.imshow( numpy.hstack((vmap[256,:,:],vmap[:,256,:],vmap[:,:,256])))
```

```python
usr = {'label':'Fisher','figsize': FS,'fontsize':FoS}

statistics.histmap( vmap, params, usr )
```
